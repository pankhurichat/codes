```python
import matplotlib.pyplot as plt
from math import *
from numpy import *
```

```python
def f(theta,v,t):
    return v
```

```python
def g(theta,v,t):
    return -4*pi**2*sin(theta)-b*f(theta,v,t)+A*sin(k*t)
```

```python
def rk4(theta0,v0,t0,tn,h,A,b,k):
    thetas=[theta0]
    velocity=[v0]
    time=[t0]
    n=int((tn-t0)/h)
    for i in range(0,n):
        k1=h*f(theta0,v0,t0)
        l1=h*g(theta0,v0,t0)
        k2=h*f(theta0+0.5*k1,v0+0.5*l1,t0+0.5*h)
        l2=h*g(theta0+0.5*k1,v0+0.5*l1,t0+0.5*h)
        k3=h*f(theta0+0.5*k2,v0+0.5*l2,t0+0.5*h)
        l3=h*g(theta0+0.5*k2,v0+0.5*l2,t0+0.5*h)
        k4=h*f(theta0+k3,v0+l3,t0+h)
        l4=h*g(theta0+k3,v0+l3,t0+h)
        
        theta0+=(1/6)*(k1+2*k2+2*k3+k4)
        v0+=(1/6)*(l1+2*l2+2*l3+l4)
        t0+=h
        
        thetas.append(theta0)
        velocity.append(v0)
        time.append(t0)
    return thetas, velocity, time
```

<b> 
    <h1>
    No driving force, 
    
 No damping pure shm 
    
</b>

```python
k=4*pi/3
A=0 #driving force
b=0 #damping coeff
theta0=1 
v0=0
t0=0
h=0.001
tn=10
x,v,t=rk4(theta0,v0,t0,tn,h,A,b,k)
plt.figure(figsize=(25,6)) 
plt.subplot(1,3,1)
plt.grid()
plt.xlabel("Thetas") ; plt.ylabel("velocity") ; plt.title("Theta vs velocity")
plt.plot(x,v)     

plt.subplot(1,3,2)
plt.plot(t,x)
plt.grid()
plt.xlabel("time") ; plt.ylabel("Thetas") ; plt.title("Theta vs Time")

plt.subplot(1,3,3)
plt.plot(t,v)
plt.grid()
plt.xlabel("time") ; plt.ylabel("Velocity") ; plt.title("Velocity vs Time")
plt.show()
```

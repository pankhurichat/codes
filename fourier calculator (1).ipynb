<h1>
    <u> Abstract </u>
    </h1>
<p>
    A Fourier series is an expansion of a periodic function f(x) in terms of an infinite sum of sines and cosines. Fourier Series makes use of the orthogonality relationships of the sine and cosine functions. A Fourier series calculator is a time saving tool which gives the series plot of any user defined function. 
    
<p> 
    Along with that we can express any periodic function as Legendre polynomial by finding the coefficients using the basis set in the orthogonal range $-1 < x < 1$ and plot the polynomial obtained
    

<h1>
    <u> Introduction </u>
    </h1>

<b>
    <h1>Fourier series:</h1>
    </b>
    <p>
    Any function following the <em> Dirichlet conditions, </em> can be represented as a sine or cosine series by finding the Euler's coefficients. The new series in whole is called the fourier series.
    $$
    a_0=\frac{1}{\pi}\int_{-L}^{L} f(x) dx
    $$
    $$ a_n=\frac{1}{\pi}\int_{-L}^{L} f(x) \cos(\frac{n(\pi)x}{L}) dx $$
    $$ b_n=\frac{1}{\pi}\int_{-L}^{L} f(x) \sin(\frac{n(\pi)x}{L}) dx $$
    
   <p>
       Once we have found the Euler Coefficients, one can easily get the fourier series using:
       $$f(x)=\frac{a_0}{2}+\sum_{n=1}^{n=\infty}[a_n\cos(\frac{n(\pi)x}{L})+b_n\sin(\frac{n(\pi)x}{L})]$$

<h1> Legendre Polynomials </h1>
<p>
    These Polynomials arise from the Legendre Differential equation:
    $$(1-x^2)y"-2xy'+n(n+1)y=0, n\in I$$
<p>
    On solving this equation we get two power series solution and on fixing the value of n, eithee of our solution reduces to a Polynomial solution.
    These are called the Legendre Polynomials:
    $$P_0(x)=1$$
    $$P_1(x)=x$$
    $$P_2(x)=\frac{1}{2}(3x^2-1)$$ and so on

<p>
    These polynomials form a basis set that can be used for defining any function, such that:
    $$f(x)=\sum_{k=0}^{\infty} A_k P_k$$
    
    Where, like Euler coefficients, 
$$  A_k=\frac{2k+1}{2} \int_{-1}^{1} P_k f(x) dx $$
    
<p>
    Thus, by finding $A_k$, we can find the function expressed as legendre polynomial

```python
from sympy import *                                                         
from math import *
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib qt
x,n=symbols("x,n")
a=str(input("Enter your function in x: "))                                 # Enter the periodic function whose series needs to be calculated 
b=(input("Enter lower limit: "))                                           # Enter the lower limit of integration
c=(input("Enter upper limit: "))                                           # Enter the upper limit of integration

l_lower=eval(str(b))
l_upper=eval(str(c))
l=(l_upper-l_lower)/2                                                      # Gives the half of periodicity L
x,y,n=symbols('x y n ')                                                    # to use x y n as symbols
fx=eval(str(a))                                                            # To evaluate the user defined function
an=[]
bn=[]
fourier=[]                                                                 # Creating empty lists to append the fourier coefficients
for n in range(1,10):                                                      # To find the first 9 terms of series
    a0=integrate(fx,(x,l_lower,l_upper))/l
    an.append((integrate(fx*(cos(n*x)),(x,l_lower,l_upper))/l)*cos(n*x))   # calculating second fourier coefficient
    bn.append((integrate(fx*(sin(n*x)),(x,l_lower,l_upper))/l)*sin(n*x))   # calculating third fourier coefficient
fourier=sum(an+bn)+a0/2                                                    # Summing the terms to get the whole fourier series
x,n=symbols('x,n')
c=[]
xx=[]
pup=[-pi,pi]                                                               # The range in which function is orthogonal
for x in range(-3,3):
    c.append(eval(str(fourier)))
    xx.append(x)
print("an is: ", an)                                                       # Printing fourier coefficients
print("bn is: ", bn)
print("a0 is: ", a0)
print("The fourier expanision is: ", fourier)
plt.figure(figsize=(18,6))                                                 # Plotting the graph of fourier series
plt.subplot(1,2,1)
plt.scatter(xx,c,label="Fourier graph",color="black")
plt.legend(loc="upper left")
plt.grid()
   
plt.subplot(1,2,2)
q=[]
p=[]
q1=[]
pup=[-pi,pi]
for x in range(-3,3):
    q.append(eval(str(a)))
    p.append(x)
    q1.append(-eval(str(a)))
plt.scatter(p,q,label="f(x)")
#plt.plot(p,q1,label="f(-x)")
plt.legend(loc="upper center")
plt.xlabel("X")
plt.ylabel("f(x)")
# plt.ylim(-100,100)
plt.grid()
plt.show()
```

```python
x=symbols('x')
# Legendre polynomial
def P(n, x):
    if(n == 0):
        return 1 # P0 = 1
    elif(n == 1):
        return x # P1 = x
    else:
        return (((2 * n)-1)*x * P(n-1, x)-(n-1)*P(n-2, x))/float(n) #bonnets recurrence relation
y=[]
X=[]
for i in range(-10,10):
    for n in range(0,10): 
        y.append(P(n,i))
        X.append(i)
l=input("Enter polynomial in x: ")

deg=int(input("Enter degree of your polynomial: "))
print("The legendre polynomial of degree", str(deg) + " is: ", P(deg,x))
x=symbols('x')
eqn=eval(str(l))
print("Your given polynomial is: ", eqn)
def a(k):
    return ((2*k+1)/2)*integrate((P(k,x)*eqn),(x,-1,1))
f_x=0
k=deg
for i in range(0,k+1):
    f_x=a(k)*P(k,x)
print("The legendre fitted polynomial is: ", f_x)
y1=[]
x1=[]
for x in range(-50,50):
    y1.append(eval(str(eqn)))
    x1.append(x)
fx=[]
for x in range(-50,50):
    fx.append(eval(str(f_x)))
x=symbols('x')
plt.plot(fx, color="red",label="Fitted poly")
plt.plot(y1,color="black",label="P(deg,x)")
plt.legend(loc="upper center")
plt.grid()
plt.show()
```
